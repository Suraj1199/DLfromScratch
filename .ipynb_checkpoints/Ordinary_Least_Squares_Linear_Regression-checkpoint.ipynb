```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import tensorflow as tf
```

```python
np.set_printoptions(threshold=10)
```

## Generating Univariate Data

```python
np.random.RandomState(seed=11)
w = np.random.randn(1) * 10
b = np.random.randn(1) * 100
noise = np.random.randn(500) * np.random.randint(0, 25, 500)
```

```python
x = np.random.permutation(np.linspace(-10, 10, 1000)[:-500])
y = w * x + b + noise
```

```python
plt.scatter(x, y)
```

## Simple Linear Regression Model

```python
class LinearRegressorOLS:
    def __init__(self):
        self.__weight = None
        self.__bias = None
    def fit(self, x, y):
        x_ = x.mean()
        y_ = y.mean()
        self.__weight = np.sum((x - x_) * (y - y_)) / np.sum((x - x_) ** 2)
        self.__bias = y_ - self.__weight * x_
    def prediction_line(self, x, y):
        y_pred = self.__weight * x + self.__bias
        plt.scatter(x, y, c='c', ec='k')
        plt.plot(x, y_pred, 'r', lw=3)
    def get_weights(self):
        return self.__weight, self.__bias
```

```python
model = LinearRegressorOLS()
```

```python
model.fit(x, y)
```

```python
model.prediction_line(x, y)
```

```python
model.get_weights()
```
